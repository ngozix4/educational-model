```python
import pandas as pd
import numpy as np

df = pd.read_csv("students training data.csv")

#choosing the targert for the model
y = df["Average Score"]
```

```python
df.describe()
```

```python
r = []
for i in df.Resources:
    if i == "Good":
        r.append(1)
        
    else:
        r.append(0)

df.Resources = r


df = df.replace(0, np.nan)
```

```python
#choosing features to train the model
features = df.drop(columns = ["Unnamed: 0",
                                   "Student ID", "Name",
                                   "Average Score"])
features = features.columns

X = df[features]
```

```python
categorical_features = ['Grade Level']
numerical_features = X.columns.difference(categorical_features)

# One-hot encode categorical features
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)
```

```python
#import sklearn library and using the DecisionTree model

from sklearn.tree import DecisionTreeRegressor

student_model = DecisionTreeRegressor(random_state = 1)

student_model.fit(X, y)

student_model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error

mask = y != 0 

X = X[mask]
y = y[mask]
predicted_scores = student_model.predict(X)
mae = mean_absolute_error(y, predicted_scores)
print(f"Mean Absolute Error: {mae}")
```

```python
from sklearn.model_selection import train_test_split

# Assume X and y are predefined
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Train a RandomForestRegressor
model = DecisionTreeRegressor()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```

```python
#training the model with RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer


imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Create and train the model
forest_model = RandomForestRegressor(n_estimators=100, random_state=42)
forest_model.fit(X_train_imputed, y_train)

# Make predictions
student_marks_pred = forest_model.predict(X_test_imputed)
forest_model = RandomForestRegressor(n_estimators=100, random_state=42)

forest_model.fit(X_train_imputed, y_train)

student_marks_pred = forest_model.predict(X_test_imputed)
print(mean_absolute_error(y_test, student_marks_pred))
```

```python
#saving the dataset
import pickle

data = "Student_Performance_Model.sav"
pickle.dump(forest_model, open(data, "wb"))

load_model = pickle.load(open(data, "rb"))
```
